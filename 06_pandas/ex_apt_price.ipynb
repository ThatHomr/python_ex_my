```python
# melt를 이용하여 월별데이터를 '기간'칼럼으로 가격은 '평당분양가격'으로
first_df_melt = first_df.melt(id_vars='지역', var_name='기간', value_vars='평당분양가격', value_name='평당분양가격')
first_df_melt.head()
```

```python
# 기간 칼럼의 값을 '연도','월'로 분리
first_df_melt['연도']=first_df_melt['기간'].str.split('년').str.get(0).astype(int)
first_df_melt['월']=first_df_melt['기간'].str.split('년').str.get(1).str.replace('월','').astype(int)
```

```python
last_df.info()
```

```python
# last_df 전용면적에서 전체만 가지고 last_df_prepare 로 저장
last_df_prepare = last_df[last_df['전용면적'] == '전체'].copy()
```

```python
last_df_prepare.head(2)
```

```python
first_df_melt.head(2)
```

```python
last_df_prepare.rename(columns={'지역명':'지역','지역':'지역명'}, inplace=True)
```

```python
# concat으로 데이터 합치기
first= first_df_melt[['지역', '연도','월','평당분양가격']]
last = last_df_prepare[['지역','연도','월','평당분양가격']]
```

```python
df = pd.concat([first, last])
```
