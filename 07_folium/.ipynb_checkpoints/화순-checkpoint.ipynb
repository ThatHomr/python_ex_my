```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
```

```python
data = pd.read_csv('202302_202302_주민등록인구및세대현황_월간.csv', encoding='CP949')
```

```python
data['행정구역'] = data['행정구역'].str.split('(').str.get(0).str.strip()
```

```python
data.columns = ['행정구역', '총인구수', '세대수', '세대당인구', '남자', '여자', '남녀비율']
```

```python
data = data.drop(columns=['남녀비율'])
```

```python
data['시'] = data['행정구역'].str.split().str.get(0).str.strip()
data['구'] = data['행정구역'].str.split().str.get(1).str.strip()
data['동'] = data['행정구역'].str.split().str.get(2).str.strip()
```

```python
hwasun = data[(data['구'] == '화순군')].reset_index(drop=True)
hwasun.head()
```

```python
hwasun = hwasun.dropna().reset_index(drop=True)
hwasun.head(15)
```

```python
hwasun['총인구수'] = hwasun['총인구수'].str.replace(',', '').astype(int)
hwasun['남자'] = hwasun['남자'].str.replace(',', '').astype(int)
hwasun['여자'] = hwasun['여자'].str.replace(',', '').astype(int)
hwasun.info()
```

```python
hwasun_chart = hwasun.groupby('동')['총인구수'].sum()
hwasun_chart =hwasun_chart.to_frame().reset_index()
hwasun_chart = hwasun_chart.sort_values('총인구수', ascending = False).reset_index(drop=True).head()
hwasun_chart = hwasun_chart.set_index('동')
hwasun_chart
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 10
plt.rcParams['axes.unicode_minus'] = False
plt.figure()
ax = plt.gca()
hwasun_chart.plot(kind='bar', ax = ax, figsize = (10,5))
```

```python
import folium,json
```

```python
jsonfile = open('HangJeongDong_ver20230101.geojson', 'r', encoding='utf8').read()
jsondata = json.loads(jsonfile)
# jsondata
jsondata_hwasun = {'type': 'FeatureCollection'}
jsondata_pick = []
jsondata_dong = []
for idx in jsondata['features']:
    if idx['properties']['sggnm'] == '화순군':
        dong = idx['properties']['adm_nm'].split()[-1].strip()
        # print(dong)
        jsondata_dong.append(dong)
        idx['id'] = dong
        jsondata_pick.append(idx)
jsondata_hwasun['features']=jsondata_pick
```

```python
map = folium.Map((35.0576218,126.9882892), zoom_start=10)
folium.Choropleth(geo_data=jsondata_hwasun,
                 data=hwasun,
                columns=['동','총인구수'],
                key_on='feature.id',
                legend_name='화순 인구수').add_to(map)
map
```
